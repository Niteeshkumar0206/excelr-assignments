```python
statistics = df.describe()
statistics.loc['mode'] = df.mode().iloc[0]
print(statistics)
```

```python
sns.histplot(df['LB'], bins=20, kde=True)
plt.xlabel('Baseline Fetal Heart Rate (FHR)')
plt.ylabel('Frequency')
plt.title('Distribution of Baseline FHR')
plt.show()
```

```python
sns.boxplot(x='AC', data=df)
plt.xlabel('Accelerations')
plt.title('Distribution of Accelerations')
plt.show()
```

```python
categorical_columns = ['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV']
for col in categorical_columns:
    plt.figure(figsize=(6, 6))
    df[col].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140)
    plt.title(col)
    plt.axis('equal')
    plt.show()
```

```python
sns.pairplot(df.sample(100))
plt.show()
```

```python
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
